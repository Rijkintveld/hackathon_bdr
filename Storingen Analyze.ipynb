```python
import pandas as pd
from datetime import date,time
```

```python
df = pd.read_csv("data_samples\storingen_ms.csv")

df.head()
```

```python
df['MTPJaar'] = df['Tijdstip_monteur_ter_plaatse'].str[0:4]
df['MTPMaand'] = df['Tijdstip_monteur_ter_plaatse'].str[5:7]
df['MTPDag'] = df['Tijdstip_monteur_ter_plaatse'].str[8:10]
df['MTPUur'] = df['Tijdstip_monteur_ter_plaatse'].str[11:13]
df['MTPMinuut'] = df['Tijdstip_monteur_ter_plaatse'].str[14:16]
df['MTPSeconde'] = df['Tijdstip_monteur_ter_plaatse'].str[17:19]

df['MNKJaar'] = df['Tijdstip_monteur_naar_klant'].str[0:4]
df['MNKMaand'] = df['Tijdstip_monteur_naar_klant'].str[5:7]
df['MNKDag'] = df['Tijdstip_monteur_naar_klant'].str[8:10]
df['MNKUur'] = df['Tijdstip_monteur_naar_klant'].str[11:13]
df['MNKMinuut'] = df['Tijdstip_monteur_naar_klant'].str[14:16]
df['MNKSeconde'] = df['Tijdstip_monteur_naar_klant'].str[17:19]

df['BSJaar'] = df['Tijdstip_begin_storing'].str[0:4]
df['BSMaand'] = df['Tijdstip_begin_storing'].str[5:7]
df['BSDag'] = df['Tijdstip_begin_storing'].str[8:10]

df.head()
```

```python
modelinput = df

modelinput['veld'] = modelinput['Naam_Onderstation'] + " " + modelinput['Veldnummer_MS-Streng']

modelinput.head()
```

```python
#Clean up
modelinput["Naam_Onderstation"] = modelinput["Naam_Onderstation"].str.lower()
modelinput['Tijdstip_begin_storing'] = pd.to_datetime(modelinput['Tijdstip_begin_storing'])
#modelinput['Reparatieduur'] = pd.to_datetime(modelinput['Reparatieduur'])
modelinput['Laatste Storing'] = pd.to_datetime('2017-11-27') - modelinput['Tijdstip_begin_storing'] 

modelinput.head()
```

```python
counts = modelinput.groupby(["Naam_Onderstation"]).size()#.reset_index()
counts.head()
final = counts.to_frame()

final.columns = ['counts']

final.head()
```

```python
final['Laatste Storing'] = modelinput['Laatste Storing'].groupby(modelinput["Naam_Onderstation"]).min()
final['Gemeentecode'] = modelinput['Gemeentecode'].groupby(modelinput["Naam_Onderstation"]).max()

final.head()
```
